1. (Gaussian Naïve Bayes Classifier) Implement Gaussian Naïve Bayes
Classifier on the Iris dataset from sklearn.datasets using
(i) Step-by-step implementation

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
iris=load_iris()
X=iris.data
y=iris.target
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42);

class GaussianNaiveBayes:
    def fit(self, X, y):
        self.classes = np.unique(y)
        self.means = {}
        self.vars = {}
        self.priors = {}

        for cls in self.classes:
            X_c = X[y == cls]
            self.means[cls] = X_c.mean(axis=0)
            self.vars[cls] = X_c.var(axis=0) + 1e-9
            self.priors[cls] = X_c.shape[0] / X.shape[0]

    def gaussian_pdf(self, x, mean, var):
        return (1 / np.sqrt(2 * np.pi * var)) * np.exp(-(x - mean)**2 / (2 * var))

    def predict(self, X):
        y_pred = []
        for x in X:
            posteriors = []
            for cls in self.classes:
                prior = np.log(self.priors[cls])
                conditional = np.sum(np.log(self.gaussian_pdf(x, self.means[cls], self.vars[cls])))
                posteriors.append(prior + conditional)
            y_pred.append(np.argmax(posteriors))
        return np.array(y_pred)

gnb = GaussianNaiveBayes()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
print("Step-by-step gaussian naive bayes accuracy:",accuracy_score(y_test, y_pred))
```

(ii) In-built function

```python

from sklearn.naive_bayes import GaussianNB

iris = load_iris()
X = iris.data
y = iris.target
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42);
model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("In-built function gaussian naive bayes accuracy:",accuracy_score(y_test, y_pred))
```

2. Explore about GridSearchCV toot in scikit-learn. This is a tool that is
often used for tuning hyperparameters of machine learning models. Use
this tool to find the best value of K for K-NN Classifier using any dataset.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
param_grid = {
    'n_neighbors': list(range(1, 31))
}
grid=GridSearchCV(knn, param_grid, cv=5)
grid.fit(X_train, y_train)
print("Best K:", grid.best_params_['n_neighbors'])
print("Best cross-validation accuracy:", grid.best_score_)
best_model=grid.best_estimator_
y_pred=best_model.predict(X_test)
print("Test accuracy:", accuracy_score(y_test, y_pred))
```
