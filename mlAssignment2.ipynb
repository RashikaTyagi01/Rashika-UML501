```python
import pandas as pd
import numpy as np
from datetime import datetime

#(a)
data = pd.read_csv('/AWCustomers.csv')
print(data.head())
print()
sales = pd.read_csv('/AWSales.csv')
print(sales.head())
data['BirthDate'] = pd.to_datetime(data['BirthDate'])
data['Age'] = datetime.now().year - data['BirthDate'].dt.year
```

```python
#(b)
merged_df = pd.merge(data,sales,on='CustomerID',how='inner')
selected=['Gender','Age','MaritalStatus','Occupation','NumberCarsOwned','NumberChildrenAtHome','TotalChildren','YearlyIncome','BikeBuyer']
df = merged_df[selected]
print(df.head())
```

Part II: Data Preprocessing and Transformation
Depending on the data type of each attribute, transform each object from your preprocessed data.
Use all the data rows (~= 18000 rows) with the selected features as input to apply all the tasks below, do not perform each task on the smaller data set that you got from your random sampling result.
(a) Handling Null values
(b) Normalization
(c) Discretization(Binning)onContinuousattributesorCategoricalAttributeswithtoomanydifferent
values
(d) Standardization/Normalization
(e) Binarization (One Hot Encoding)

```python
#(a) Handling Null values
df.isnull().sum()
df=df.dropna()
```

```python
#(b) Normalization
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
cols = ['Age','NumberCarsOwned','NumberChildrenAtHome','TotalChildren','YearlyIncome']
df[cols] = scaler.fit_transform(df[cols])
print(df.head())
```

```python
#(c) Discretization(Binning)
df['AgeGroup'] = pd.cut(df['Age'],bins=3,labels=["Young","Middle","Old"])
df['IncomeGroup'] = pd.cut(df['YearlyIncome'],bins=3,labels=["Low","Medium","High"])
print(df[['AgeGroup','IncomeGroup']].head())
```

```python
#(d) Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numeric_col = ['Age','NumberCarsOwned','NumberChildrenAtHome','TotalChildren','YearlyIncome']
df[numeric_col] = scaler.fit_transform(df[numeric_col])
print(df.head())
```

```python
#(e) Binarization (One Hot Encoding)
pd.get_dummies(df,columns=['Gender','MaritalStatus','Occupation'])
print(df.head())
```

Part III: Calculating Proximity /Correlation Analysis of two features
Make sure each attribute is transformed in a same scale for numeric attributes and Binarization for each nominal attribute, and each discretized numeric attribute to standardization. Make sure to apply a correct similarity measure for nominal (one hot encoding)/binary attributes and numeric attributes respectively.
(a) Calculate Similarity in Simple Matching, Jaccard Similarity, and Cosine Similarity between two following objects of your transformed input data.
(b) Calculate Correlation between two features Commute Distance and Yearly Income

```python
#(a)
from sklearn.metrics.pairwise import cosine_similarity

df_encoded = pd.get_dummies(df, drop_first=True)
obj1 = df_encoded.iloc[0].values
obj2 = df_encoded.iloc[1].values


def smc(x,y):
  return np.sum(x==y)/len(x)

def jaccard(x,y):
  intersection=np.sum((x==1) & (y==1))
  union = np.sum((x==1) | (y==1))
  if union!=0:
    return intersection/union
  else:
    return 0

print(smc(obj1,obj2))
print(jaccard(obj1,obj2))
print(cosine_similarity([obj1],[obj2])[0][0])


```
